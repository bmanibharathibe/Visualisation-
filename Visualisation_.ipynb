```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
tips = sns.load_dataset("tips")

# -------------------------------
# Task 1: Matplotlib Bar Chart
# -------------------------------
tips_sum = tips.groupby("day")["tip"].sum()

plt.figure(figsize=(8, 5))
plt.bar(tips_sum.index, tips_sum.values)
plt.title("Total Tips Collected per Day")
plt.xlabel("Day")
plt.ylabel("Total Tip Amount")
plt.show()


# -------------------------------
# Task 2: Seaborn Histogram
# -------------------------------
plt.figure(figsize=(8, 5))
sns.histplot(tips["tip"], bins=20, kde=True)
plt.title("Distribution of Tip Amounts")
plt.show()


# -------------------------------
# Task 3: Seaborn Scatter Plot
# -------------------------------
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Total Bill vs Tip Amount")
plt.show()
```
